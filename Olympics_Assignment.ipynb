# **Importing Libraries**

```python
import pandas as pd
import plotly.express as px
```

# **Importing Data**

```python
# 1. Load the CSV
df = pd.read_csv("athlete_events.csv")

# Quick look
df.head()
```

```python
print(df.isna().sum())    # missing values per column
```

# **Data Cleaning**

```python
# Critical columns must not be null
critical_cols = ["Name", "Sex", "Age", "Year", "Sport", "Event"]

# Drop rows where any critical column is NaN
df = df.dropna(subset=critical_cols)

# Impute non-critical numeric columns with median
for col in ["Height", "Weight"]:
    if col in df.columns:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)

# For Medal, fill missing with the string "None"
if "Medal" in df.columns:
    df["Medal"] = df["Medal"].fillna("None")

# Check remaining missing values
print("Missing values after cleaning:")
print(df.isna().sum())
```

```python
# Convert Year to datetime (year-only)
df["Year"] = pd.to_datetime(df["Year"].astype(int), format="%Y", errors="coerce")

# Optionally convert Age to integer if currently float
if pd.api.types.is_float_dtype(df["Age"]):
    df["Age"] = df["Age"].astype("Int64")  # nullable integer

print(df.dtypes)
```

```python
# Quick look
df.head(20)
```

```python
# 5. Remove exact duplicate rows
before_dups = len(df)
df = df.drop_duplicates()
after_dups = len(df)

print(f"Removed {before_dups - after_dups} duplicate rows")
```

# **Data wrangling**

```python
# Create Age_Group column with custom bins
age_bins = [0, 18, 25, 35, 100]
age_labels = ['0-18', '19-25', '26-35', '36+']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=True)
```

```python
df.head()
```

```python
df['Age_Group'].value_counts().sort_index()
```

```python
# Create Century column from Year
df['Century'] = (df['Year'].dt.year // 100 + 1).astype(str) + 'th'
```

# **Data Analysis**

```python
avg_age_per_event = df.groupby("Event")["Age"].mean().reset_index()
avg_age_per_event.head()
```

```python
athlete_medal_counts = (
    df[df["Medal"].notna()]                 # count only medal-winning entries
    .groupby(["Sport", "Name"])
    .size()
    .reset_index(name="Medal_Count")
)

# For each sport, keep the athlete with the max medals
top_athlete_per_sport = (
    athlete_medal_counts
    .sort_values(["Sport", "Medal_Count"], ascending=[True, False])
    .groupby("Sport")
    .first()
    .reset_index()
)

top_athlete_per_sport
```

```python
gold_medals = (
    df[df["Medal"] == "Gold"]
    .groupby("Team")
    .size()
    .sort_values(ascending=False)
    .head(10)
    .reset_index(name="Gold_Count")
)

gold_medals
```

# **Data Visualisation**

```python
gold = (
    df[df["Medal"] == "Gold"]
    .groupby("Team")
    .size()
    .sort_values(ascending=False)
    .head(10)
    .reset_index(name="Gold_Count")
)

fig = px.choropleth(
    gold,
    locations="Team",
    locationmode="country names",
    color="Gold_Count",
    hover_name="Team",
    color_continuous_scale="sunset",
)

fig.update_geos(
    projection_type="orthographic",   # ROUND GLOBE
    showcoastlines=True,
    showcountries=True,
    showland=True,
    landcolor="lightgray",
    showocean=True,
    oceancolor="lightblue"
)

fig.update_layout(
    title="Countries and their Gold medal counts",
    template="plotly_white",
    height=700,
    paper_bgcolor="white",  # make the background blue too
)

fig
```

```python
participation = (
    df.groupby(["Year", "Team"])
    .size()
    .reset_index(name="Athlete_Count")
)

fig = px.choropleth(
    participation,
    locations="Team",
    locationmode="country names",
    color="Athlete_Count",
    hover_name="Team",
    animation_frame="Year",
    color_continuous_scale="Plasma",
    title="Global Olympic Participation Over Time"
)

fig.update_layout(template="plotly_dark")

fig
```
